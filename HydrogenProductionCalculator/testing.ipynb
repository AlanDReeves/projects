```python
from productionCalculator import productionCalculator as PC
calc = PC()
calc.readFile("curve1.txt")
```

```python
print(calc.firstDerivative(1, 2))
print(calc.secondDerivative(2, 3))
```

```python
firstDerivs = []
for i in range(1, len(calc.voltages)):
    firstDerivs.append(calc.firstDerivative(i-1, i))

firstDerivs.append(float(0))

secondDerivs = []
for i in range(1, len(calc.voltages)):
    secondDerivs.append(calc.secondDerivative(i-1, i))

secondDerivs.append(0)
```

```python
import matplotlib.pyplot as plt
plt.plot(calc.voltages, firstDerivs)
plt.xlabel("voltage (v)")
plt.ylabel("change in voltage")
plt.show()

plt.plot(calc.voltages, secondDerivs)
plt.xlabel("voltage (v)")
plt.ylabel("change in derivative")
plt.show()
```
